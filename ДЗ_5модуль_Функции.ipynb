Примеры работы:

```python
documents = [
              {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'}, 
              {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'}, 
              {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'} 
]
directories = {
                '1': ['2207 876234', '11-2'],
                '2': ['10006'],
                '3': []
}
def adding_shelf(doc):
    doc = directories
    shelf_number = input('Введите номер полки: ')
    for key in doc.keys():
        if key == shelf_number:
            print('Такая полка уже существует. Текущий перечень полок: ' + ','.join(doc.keys()))
            break
    if shelf_number not in doc.keys():
        doc.setdefault(shelf_number, [])
        print('Полка добавлена. Текущий перечень полок: ' + ','.join(doc.keys()))
```

```python
documents = [
              {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'}, 
              {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'}, 
              {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'} 
]
directories = {
                '1': ['2207 876234', '11-2'],
                '2': ['10006'],
                '3': []
}
def doc_owner(doc): 
    num = input('Введите номер документа: ')
    count = 0  
    for docum in doc:
        if docum['number'] == num:
            print("Результат:")
            print("Владелец документа:", docum['name'])
        elif docum['number'] != num:
            count += 1
            if count == len(doc):
                print("Результат:")
                print("Документ не найден в базе")
def doc_st_shelf(doc):
    doc = directories
    num = input('Введите номер документа: ')
    count = 0 
    for folder, docum in doc.items():
        if num in docum:
            print('Результат:')
            print(f'Документ хранится на полке: {folder}')
        else:
            count += 1
            if count == len(doc):
                print("Результат:")
                print("Документ не найден в базе")
def get_key(doc, val):
    for key, value in doc.items():
        if val in value:
            return key
def inf_all_doc(doc):
    for el in doc:
        doc_type = el['type']
        number = el['number']
        name = el['name']
        print('Результат:')
        print('№: {1}, тип: {0}, владелец: {2} полка хранения: {3}'.format(doc_type, number, name, get_key(directories, number)))
def adding_shelf(doc):
    doc = directories
    shelf_number = input('Введите номер полки: ')
    for key in doc.keys():
        if key == shelf_number:
            print('Такая полка уже существует. Текущий перечень полок: ' + ','.join(doc.keys()))
            break
    if shelf_number not in doc.keys():
        doc.setdefault(shelf_number, [])
        print('Полка добавлена. Текущий перечень полок: ' + ','.join(doc.keys()))
def remov_shelf(doc):
    doc = directories
    shelf_number = input('Введите номер полки: ')
    if shelf_number not in doc.keys():
        print('Такой полки не существует. Текущий перечень полок: ' + ','.join(doc.keys()))
    for key, value in list(doc.items()):
        if key == shelf_number and value != []:
            print('На полке есть документы, удалите их перед удалением полки. Текущий перечень полок: ' + ','.join(doc.keys()))
        elif key == shelf_number and value == []:
            del doc[shelf_number]
            print('Полка удалена. Текущий перечень полок: ' + ','.join(doc.keys()))
def main(doc):
    while True:
        user_cmd = input('Введите команду: ')
        if user_cmd == 'p':
            doc_owner(doc)
        elif user_cmd == 's':
            doc_st_shelf(doc)
        elif user_cmd == 'l':
            inf_all_doc(doc)
        elif user_cmd == 'as':
            adding_shelf(doc) 
        elif user_cmd == 'ds':
            remov_shelf(doc)
        elif user_cmd == 'z':
            print("До свидания!")
            break
main(documents)           
```
