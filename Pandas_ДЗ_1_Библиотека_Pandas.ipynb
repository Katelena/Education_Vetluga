```python
import pandas as pd
```

```python
data = pd.read_csv('ratings.csv')
data.head()
```

```python
max_rating = data['rating'].max() # ввели переменную для рейтинга "5"
data = data[['movieId', 'rating']] # выбрали необходимые столбцы
data = data[data.rating == max_rating] # отсортировали строки с рейтингом "5"
data.groupby('movieId').count().sort_values('rating', ascending=False).head(1)  
# сгруппировали по фильмам,посчитали количество оценок по каждому из них, отсортировали по убыванию, отразили строку с максимальным количеством
```

#### Задание 2
По данным файла power.csv посчитайте суммарное потребление стран Прибалтики (Латвия, Литва и Эстония) категорий 4, 12 и 21 за период с 2005 по 2010 года. Не учитывайте в расчетах отрицательные значения quantity.


```python
countries = pd.read_csv('power.csv')
countries.head()
```

```python
filtered_countries = countries[(countries['country'] == 'Lithuania') | 
                               (countries['country'] == 'Latvia') | 
                               (countries['country'] == 'Estonia')]
filtered_category = filtered_countries[(filtered_countries['category'] == 4) | 
                                       (filtered_countries['category'] == 12) | 
                                       (filtered_countries['category'] == 21)]
filtered_quantity = filtered_category[filtered_category['quantity'] > 0]
filtered_year = filtered_quantity[(filtered_quantity['year'] >= 2005) & (filtered_quantity['year'] <= 2010)]
filtered_columns = filtered_year[['country', 'quantity']]
filtered_columns.groupby('country').sum().head() # Сгруппировали по странам и посчитали сумму по каждй из них
```
